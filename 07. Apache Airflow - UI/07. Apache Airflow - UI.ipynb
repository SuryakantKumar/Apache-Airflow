# <span style="color:skyblue">**Apache Airflow - UI**</span>

By understanding and utilizing these views and functionalities, We can effectively `monitor`, `manage`, and `troubleshoot` our data pipelines within Apache Airflow. 

1. **DAGs View**

    - Provides an overview of all DAGs present in the system.

    - Features a `toggle` to `pause` or `unpause` DAG execution.

    - Displays DAG `name`, `tags`, `owner`, `schedule interval`, `last run`, and `task statuses`.
    
    - Allows actions like triggering a DAG `manually`, `refreshing`, and `deleting` (Only deletes the meta database not the code).
    
    

2. **Tree View**

    - Presents a hierarchical representation of DAG runs and task instances.

    - Each square represents a task, and its color indicates its status (success, running, failed, etc.).

    - Useful for visualizing the history and status of DAG runs and tasks.
    
    
    
3. **Graph View**

    - Offers a graphical representation of DAG structure and task dependencies.

    - Enables users to quickly visualize `task dependencies` and `execution order`.

    - Provides `auto-refresh` functionality to monitor task execution in `real-time`.

    

4. **Gantt View**

    - Shows task durations and dependencies in a Gantt chart format.

    - Useful for identifying task bottlenecks and parallelizable tasks.

    - Longer rectangles indicate tasks taking more time to complete.
    
    

5. **Code View**

    - Allows users to review the code of a DAG to ensure modifications are applied correctly.

    - Ensures that Airflow is aware of the changes made to DAG definitions.

    

6. **Context Menu**

    - Accessed by clicking on a task instance in the Graph or Tree view.

    - Provides options to view instance details, logs, and task actions (retry, clear, modify status).
    
    

7. **Task Actions:**

    - Offers actions like retrying tasks, modifying task status, and clearing tasks.

    - Allows users to manipulate task instances and test different scenarios.
