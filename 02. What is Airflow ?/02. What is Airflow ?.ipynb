# <span style="color:skyblue">**What is Airflow ?**</span>

Apache `Airflow` is an open-source orchestration tool that programmatically `create`, `schedule` and `monitor` workflows.

### <span style="color:orange">Benefits of Using Airflow</span>

* **Dynamic:** Everything is coded in Python, allowing easy creation and customization of tasks and data pipelines.

* **Scalable:** It can handle various tasks based on available architecture, resources, or budget. Depending on the architechture, Airflow has integrations for popular tools and services such as AWS, GCP, Azure, Kubernetes, Docker, Spark, Kafka, and Hive executors to run tasks.

* **User Interface (UI):** It provides a user-friendly UI for monitoring tasks and pipelines.

* **Extensible:** It allows creation of custom operators and plugins. Custom operators can be written in Python and can be integrated with any external system that has a Python API or a command-line interface. It means that, Everything that we can do with Python, we can implement the same in our data pipelines. This ensures flexibility and adoption to new tools.

### <span style="color:orange">Core Components of Airflow</span>

- **Web Server:** Flask server with Gunicorn hosts the user interface. We don't need web server to run airflow but without it, we can't access the UI.

- **Scheduler:** It is essential for scheduling tasks and data pipelines. If the schedular goes down, we won't be able to run any more tasks. That's why, we should have at least 2 schedulers in production.

- **Metastore:** It stores metadata related to tasks and pipelines. It is typically a database compatible with SQL Alchemy such as PostgreSQL, MySQL or Oracle DB.

- **Triggerer:** It manages deferrable operators that can suspend and resume tasks.

- **Executor:** It is class that defines `how` and `where` tasks are executed (Example: Kubernetes, Celery). If we have one Single machine, We will use local executor to execute our tasks on single machine.

- **Worker:** It is a process or machine which execute tasks as per defined components and instructions.

### <span style="color:orange">Core Concepts in Airflow</span>

- **DAG (Directed Acyclic Graph):** It represents the data pipeline in Airflow, showcasing dependencies between tasks without any cycles (infinite loops). Here, We can see task (T4) depends on tasks (T1), (T2) and (T3) and there is no cycle in this data pipeline. But In the second pipeline there is a cycle so, It's not a DAG.



- **Operator:** Whenever we want to create a task in airflow, we use an operator. An Operator encapsulates the logic that we want to achieve in the task.

    - **Action Operators:** We can execute Python function using `python` operator, Bash script with a `bash` operator, SQL request with `postgres` operator.

    - **Transfer Operators:** If we want to transfer data between source `A` to destination `B` then we can use `transfer` operator (`mysql` to `postgres` operator).

    - **Sensor Operators:** It waits for an event (like database entry or file landing) to happen before triggering the next task.

    - **Deferrable Operators:** These are enhanced sensors waiting for specific events before proceeding.

As soon as We trigger a task with these operators, that task becomes a `task instance`. Basically, A Task instance is an instance of an operator with a specific date (run time).

### <span style="color:orange">What Airflow is NOT</span>

- **Not a Data Processing Framework:** It's not ideal for directly processing data but serves as a trigger for data processing tools like Spark.

- **Not a Data Streaming Solution:** Airflow isn't meant for data streaming; Other solutions like Spark Streaming are better suited for that purpose.
