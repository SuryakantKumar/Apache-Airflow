# <span style="color:skyblue">**Apache Airflow - CLI**</span>

With the Apache `Airflow` command-line interface (`CLI`), We can interact with a lot of different processes such as our `dags`, `task instances`, `dag runs` and so on.

### <span style="color:orange">Accessing the CLI</span>

* We can list the running docker containers using the command `docker ps`

* We can access the CLI by opening a `bash` session inside the Docker `container` of Airflow using the command **`docker exec -it <container_id> /bin/bash`**


```shell
docker exec -it 261344369fc4 /bin/bash
```



* As we can see from the prompt, we are inside the Docker container.

### <span style="color:orange">Grouping of Commands</span>

* If we want any help regarding Airflow commands, we can use **`airflow -h`**.

* Since, Commands are grouped based on the resource being interacted with, like `connections`, `dags`, `tasks`. We have to use the commands like **`airflow dags -h`** to get the commands related to our dags.

```shell
airflow -h
```



```shell
airflow dags -h
```



* Like this we can execute the commands for other groups if needed.

### <span style="color:orange">Important CLI Commands</span>

- **`airflow db init`** command initializes the metadata database and generates necessary files and folders.

- **`airflow db reset`** command resets the metadata database, removing all metadata (Use with caution, Typically not used in production).

- **`airflow db upgrade`** command upgrades database schemas and values that are in the metadata base of airflow when updating Airflow versions.

- **`airflow webserver`** command starts the web server and the user interface of airflow.

- **`airflow scheduler`** command starts the Airflow scheduler.

- **`airflow celery worker`** command starts an Airflow worker for Celery distributed task execution.

- **`airflow dags list`** command lists all DAGs recognized by Airflow (Either those in the examples folder or dags folder). This is really useful if we just added a new Dag and we want to verify if airflow is aware of it.

- **`airflow dags trigger <dag_id>`** command Manually triggers a DAG run with the current date as execution date.

- **`airflow dags trigger <dag_id> -e 2024-02-05`** command Manually triggers a DAG run with a date in the past as execution date (This won’t trigger the tasks of that dag unless we set the option `catchup=True` in the DAG definition)

- **`airflow dags trigger <dag_id> -e '2024-02-10 06:00:00+00:00'`** command Manually triggers a DAG run with a date in the future. The dag will be scheduled at that date.

- **`airflow dags list-runs -d <dag_id>`** command Lists history of given DAG runs with some information such as the run ID, the state, the execution date, start date, end date, and so on.

- **`airflow dags backfill -s 2024-02-01 -e 2024-02-05`** command Backfills DAG runs between specified start and end dates, useful for running missed DAG runs.

- **`airflow dags backfill -s 2024-02-01 -e 2024-02-05 --reset-dagruns`** command Backfills DAG runs between specified start and end dates, if we want to re-run already triggered Dag runs. (Let's imagine that we just figured out that we made a mistake in our Dag and that Dag was running during the last past five days. With `--reset-dagruns`, we will be able to rerun those Dag runs.)

- **`airflow tasks list <dag_id>`** command Lists tasks of a specific DAG. (Let's imagine that we added a new task to our Dag and we want to make sure that airflow is aware of that new task or not.)

- **`airflow tasks test <dag_id> <task_id> 2024-02-05`** command Tests a task without checking dependencies or storing metadata, ensuring the task works as expected. (Each time we add a new task to our Dag, make sure that the task works, before triggering the entire DAG run. Try running for past date and it will show that the tas has executed.)
