# <span style="color:skyblue">**Introduction To Forex Data Pipeline**</span>

We're embarking on a comprehensive journey to create our first data pipeline using Apache Airflow.

### <span style="color:orange">Goal of the Data Pipeline</span>

* We'll be processing forex data, specifically the fluctuation rates between currencies, such as the euro against the USD.



### <span style="color:orange">Flow Chart of the Data Pipeline</span>



- **Task 1:** Check if forex rates are available by accessing a specific URL.
   
- **Task 2:** Verify the availability of the file containing the currency rates.

- **Task 3:** Download the forex rates using a Python function.

- **Task 4:** Save the forex rates into HDFS (Hadoop Distributed File System).

- **Task 5:** Create a Hive table to interact with the stored forex rates.

- **Task 6:** Process the forex rates with Spark, a distributed data processing framework.

- **Task 7:** Send an email notification when the data pipeline is completed.

- **Task 8:** Send a Slack notification to inform the team about the completion of the data pipeline.

### <span style="color:orange">Architecture of the Data Pipeline</span>



- **Back End:** Includes `HDFS` for data storage, `Spark` for data processing, and `Hive` for querying data using SQL-like syntax.
   
- **Front End:** Consists of a `Postgres` database used by `Hive` and `Adminer` for database management, and `Hue` for data visualization and dashboarding.

- **Airflow:** Serves as the orchestrator for the data pipeline, enabling us to create, schedule, and monitor workflows that interact with various components of the architecture.
